# 常微分方程式の数値計算
## JuliaによるEuler法の実装例（1次元）

$f(t,u) = u^2 + 1$, $u_0 = 0$の場合を考える．
区間$[0,T]$は$N$等分されているとする．
JuliaによるEuler法の実装例は次のようになる．


```julia
f(t, u) = u^2 + 1   # 厳密解は u(t) = tan(t)
u = 0
T = pi/4
N = 10
h = T/N

for i in 0:N-1
    t = h*i
    u += h * f(t, u)
    println("u($(t+h)) = $u")
end
```

## ベクトル値関数版

```julia
function f(t, u)
    [u[1]*(1-u[2]), -u[2]*(1-u[1])]
end
T = 50; 
N= T*100; 
h=T/N;
U = Float64[1,4]    # 初期値  
Uhist = U   # 履歴保存用

for t in 0:h:T  # t=0,h,2h,...,T
    U += h*f(t,U)
    Uhist = hcat(Uhist, U)  # 履歴追加
end
```

```julia
# グラフ描画
using PyPlot 
#cla()   # キャンバスクリア
plot(Uhist[1,:],Uhist[2,:],marker=nothing)
grid()
```
